# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams.update({'font.size': 30})
```

```python
data = pd.read_csv("./ElectricCarData.csv")
```

```python
data = data.drop('Segment',axis=1)
```

```python
data.head()
```

```python
def convert_float(x):
    return(float(x[:-4]))
data['Accel'] = data['Accel'].apply(convert_float)
```

```python
def convert_float(x):
    return(float(x[:-5]))
data['TopSpeed'] = data['TopSpeed'].apply(convert_float)
```

```python
def convert_float(x):
    return(float(x[:-3]))
data['Range'] = data['Range'].apply(convert_float)
```

```python
def convert_float(x):
    return(float(x[:-6]))
data['Efficiency'] = data['Efficiency'].apply(convert_float)
```

```python
def convert_float(x):
    try:
        return(float(x[:-5]))
    except:
        return 0
data['FastCharge'] = data['FastCharge'].apply(convert_float)
```

```python
data['RapidCharge'].unique()
```

```python
data['RapidCharge']=data['RapidCharge'].replace(to_replace='Rapid charging possible',value=1)
data['RapidCharge']=data['RapidCharge'].replace(to_replace='Rapid charging not possible',value=0)
```

```python
data_bool=pd.get_dummies(data.drop(['Brand','Model'],axis=1),drop_first=False)
```

```python
data_bool
```

# PCA Analysis

```python
from sklearn.decomposition import PCA
```

```python
pd.options.display.float_format = '{:.6f}'.format
pca = PCA(n_components=11)
pca.fit(data_bool)
pca_data = pca.transform(data_bool)
df = pd.DataFrame(pca_data)
df.describe()
```

```python
loadings = pd.DataFrame(pca.components_.T, columns=df.columns, index=data_bool.columns)
loadings
```

```python
sns.scatterplot(x=pca_data[:,0],y=pca_data[:,1],s=1000,hue=data['PriceEuro'])
plt.xlabel("PC1")
plt.ylabel("PC2")
```

# Clustering The data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(data_bool)
```

```python
import warnings
warnings.filterwarnings('ignore')
X=data_bool
ssd = []
from sklearn.cluster import KMeans
for k in range(2,10):
    model = KMeans(n_clusters=k,n_init=10)
    model.fit(X)
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,10),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

```python
from sklearn.metrics import silhouette_score
silhouette_coefficients = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.plot(range(2,10),silhouette_coefficients,'o--')
plt.xlabel("K Value")
plt.ylabel("Silhouette Score")
```

```python
from sklearn.datasets import make_moons
features, true_labels = make_moons(n_samples=len(data_bool), noise=0.05)
```

```python
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import adjusted_mutual_info_score
ari = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    score = adjusted_rand_score(true_labels, kmeans.labels_)
    ari.append(score)
```

```python
plt.plot(range(2,10),ari,'o--')
plt.xlabel("K Value")
plt.ylabel("Adjusted Random Index Score")
```

```python
final_kmeans = KMeans(n_clusters=5)
final_kmeans.fit(X)
```

```python
data_bool["Cluster"]=final_kmeans.labels_
```
